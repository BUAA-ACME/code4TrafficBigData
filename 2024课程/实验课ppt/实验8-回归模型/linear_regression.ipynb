```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('21年6月实时数据NJ-S122其林门_train.csv')
df['volume'] = df[['XKC', 'DKC', 'XHC', 'ZHC', 'DHC', 'TDH']].sum(axis=1)
df = df[(df['volume']>0) & (df['PJCTJJ']>0) & (df['SJZYL']>0)]
df.head()
```

```python
X = np.log(df[['PJCTJJ', 'SJZYL']])
X = sm.add_constant(X)
y = np.log(df['volume'])
model = sm.OLS(y, X)
result = model.fit()
print(result.summary())
```

```python
import statsmodels.formula.api as smf

model = smf.ols('np.log(volume) ~ np.log(PJCTJJ) + np.log(SJZYL)', df)
result = model.fit()
print(result.summary())
```

```python
df_test = pd.read_csv('./21年6月实时数据NJ-S122其林门_test.csv')
predict = result.predict(df_test[['PJCTJJ', 'SJZYL']])
predict = np.exp(predict)
predict
```

```python
predict.to_csv('pred.csv', index=False)
```
